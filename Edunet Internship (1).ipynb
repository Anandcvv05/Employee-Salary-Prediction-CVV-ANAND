```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.utils import to_categorical

# Load the dataset
df = pd.read_csv('adult 3.csv')

# Check for nulls
df.dropna(inplace=True)

# Encode categorical variables
for column in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])

# Split features and target
X = df.drop('income', axis=1)  # replace 'class' with the actual label column name
y = df['income']  # ensure this is your target column

# One-hot encode labels (if classification)
y = to_categorical(y)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape for CNN (samples, height, width, channels)
X_cnn = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1, 1)

# Reshape for RNN (samples, timesteps, features)
X_rnn = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# Train-test split
X_train_cnn, X_test_cnn, y_train, y_test = train_test_split(X_cnn, y, test_size=0.2, random_state=42)
X_train_rnn, X_test_rnn, _, _ = train_test_split(X_rnn, y, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, Dropout

cnn_model = Sequential([
    Conv2D(32, kernel_size=(3, 1), activation='relu', input_shape=(X_cnn.shape[1], 1, 1)),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),-
    Dense(y.shape[1], activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()

cnn_model.fit(X_train_cnn, y_train, epochs=10, validation_data=(X_test_cnn, y_test), batch_size=32)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

rnn_model = Sequential([
    LSTM(64, input_shape=(X_rnn.shape[1], 1)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(y.shape[1], activation='softmax')
])

rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
rnn_model.summary()

rnn_model.fit(X_train_rnn, y_train, epochs=10, validation_data=(X_test_rnn, y_test), batch_size=32)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Predictions for CNN
y_pred_cnn = cnn_model.predict(X_test_cnn)
y_pred_classes_cnn = np.argmax(y_pred_cnn, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Predictions for RNN
y_pred_rnn = rnn_model.predict(X_test_rnn)
y_pred_classes_rnn = np.argmax(y_pred_rnn, axis=1)
```

```python
print("CNN Classification Report:")
print(classification_report(y_true_classes, y_pred_classes_cnn))

# Confusion Matrix
cm_cnn = confusion_matrix(y_true_classes, y_pred_classes_cnn)
disp_cnn = ConfusionMatrixDisplay(confusion_matrix=cm_cnn)
disp_cnn.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - CNN")
plt.show()
```

```python
print("RNN Classification Report:")
print(classification_report(y_true_classes, y_pred_classes_rnn))

# Confusion Matrix
cm_rnn = confusion_matrix(y_true_classes, y_pred_classes_rnn)
disp_rnn = ConfusionMatrixDisplay(confusion_matrix=cm_rnn)
disp_rnn.plot(cmap=plt.cm.Greens)
plt.title("Confusion Matrix - RNN")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

acc_cnn = accuracy_score(y_true_classes, y_pred_classes_cnn)
acc_rnn = accuracy_score(y_true_classes, y_pred_classes_rnn)

print(f"CNN Accuracy: {acc_cnn:.4f}")
print(f"RNN Accuracy: {acc_rnn:.4f}")
```
